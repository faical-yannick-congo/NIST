```python
from bokeh.charts import Scatter, Histogram, output_file, show
from bokeh.io import output_notebook
import bokeh.palettes as palettes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from bokeh.plotting import figure, show

output_notebook()
p = figure()

a = pd.DataFrame(np.random.normal(0,1,size=(100,2)), columns=['x','y'])
a['label'] = [i%20 for i in range(100)]

s = Scatter(a, x='x', y='y', color='label', palette=palettes.Category20[20])
show(s)
```

```python
df = pd.read_csv('dftsne_minSamp.csv')
```

```python
s = Scatter(df, x='x', y='y', color='minSample25', palette=palettes.Category20[20])
show(s)
```

```python
from bokeh.models import Legend
from bokeh.models import HoverTool, BoxSelectTool
from bokeh.plotting import figure, output_file, show, ColumnDataSource
TOOLS = [BoxSelectTool(), HoverTool()]

x = df['x']
y = df['y']
samp = df['minSample10']
source = ColumnDataSource(
        data=dict(
            x=x,
            y=y,
            samp=samp,
        )
    )

hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("(x,y)", "($x, $y)"),
            ("Group ID", "@minSample10"),
        ]
    )

s = Scatter(df, x='x', y='y', color='minSample10', palette=palettes.Category20[20],tools=[hover, 'pan', 'wheel_zoom', 'save'])


# s.legend.orientation = "horizontal"
# minS = df['minSample10']
# legend = Legend(items=minS, location=(0, -30))
# s.add_layout(legend, 'right')
show(s)
```

```python
output_file("toolbar.html")

source = ColumnDataSource(
        data=dict(
            x=[1, 2, 3, 4, 5],
            y=[2, 5, 8, 2, 7],
            desc=['A', 'b', 'C', 'd', 'E'],
        )
    )

hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("(x,y)", "($x, $y)"),
            ("desc", "@desc"),
            ("desc", "@desc"),
        ]
    )

p = figure(plot_width=400, plot_height=400, tools=[hover],
           title="Mouse over the dots")

p.circle('x', 'y', size=20, source=source)

show(p)
```
